# K-Means Clustering

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.environ["OMP_NUM_THREADS"] = '1'
from sklearn.cluster import KMeans
```

```python
relevant_cols = ["Age", "Annual Income (k$)", 
                 "Spending Score (1-100)"]

customer_df = df[relevant_cols]
```

```python
customer_df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(customer_df)
```

```python
scaled_data = scaler.transform(customer_df)
```

```python
scaled_data
```

```python
def find_best_clusters(df, maximum_K):
    clusters_centers = []
    k_values = []
    for k in range(2, maximum_K):
        kmeans_model = KMeans(n_clusters = k)
        kmeans_model.fit(df)

        clusters_centers.append(kmeans_model.inertia_)
        k_values.append(k)

    return clusters_centers, k_values
```

```python
clusters_centers, k_values = find_best_clusters(scaled_data, 12)
```

```python
def generate_elbow_plot(clusters_centers, k_values):
    
    figure = plt.subplots(figsize = (12, 6))
    plt.plot(k_values, clusters_centers, 'o-', color = 'orange')
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("Cluster Inertia")
    plt.title("Elbow Plot of KMeans")
    plt.show()
```

```python
generate_elbow_plot(clusters_centers, k_values)
```

```python
kmeans_model = KMeans(n_clusters = 5)
```

```python
kmeans_model.fit(scaled_data)
```

```python
df["clusters"] = kmeans_model.labels_
```

```python
df
```

```python
plt.scatter(df["Spending Score (1-100)"], 
            df["Annual Income (k$)"], 
            c = df["clusters"]
            )
```

# Principal Component Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
from sklearn.datasets import load_breast_cancer
cancer_data = load_breast_cancer()
print(type(cancer_data))
```

```python
cancer_data.keys()
```

```python
no_features=len(cancer_data['feature_names'])
print(no_features)
```

```python
no_records=len(cancer_data['data'])
print(no_records)
```

```python
df = pd.DataFrame(cancer_data['data'],columns=cancer_data['feature_names'])
df
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_scaled=scaler.fit_transform(df)
df_scaled
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pc12 = pca.fit_transform(df_scaled)
pc12.shape
```

```python
df_pca = pd.DataFrame(pc12,columns=['PC1','PC2'])
df_pca.head()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(
    x=df_pca['PC1'],
    y=df_pca['PC2'],
    c=cancer_data['target'],
    cmap='Set1')
plt.xlabel('First Principal Component - PC1')
plt.ylabel('Second Principal Component - PC2');
```

```python
X = df_pca
y = cancer_data['target']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
from sklearn.linear_model import LogisticRegression
logR = LogisticRegression()
logR.fit(X_train,y_train)
predictions = logR.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test,predictions))
```
